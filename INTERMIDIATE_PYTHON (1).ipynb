```python
# LISTS

# Lists are collectable datatypes that are mutable , ordered and duplicated items.

# As we recollect our learnt experience lists are commanded with square brackets with elemnts inside the brackects seperated by commas.

mylist = ["Chowder","Molemo","Batman","Thabang","Kamogelo"]



# Our lists can contain different datatypes (interger , bolean function and strings)

# It also allows us to duplicate elements.

mylist = ["Chowder","Molemo","Batman","Thabang","Kamogelo"]

print(mylist)

mylist2 = [45 , True , "Sendy" ,"Sendy"]

print(mylist2)

# To access a certain element you will need to refer to its index.

Item = mylist[3]
print(Item)

# To refer to the last item , we use "-1"

# and if you want to iterate your list you can use a "FOR in LOOP"

for i in mylist2 :
  print(i)


# So if you want to verify if an item is an list , you can use an "IF STATEMENT"

mylist = ["Chowder", "Bimbo"]
print(mylist)

for Chowder in mylist:
  print("TRUE")
else:
  print("False")


# To append/ find the length and the use of more built in functions, you use:

#Length
mylist = [45 , 4 , 56 , 3]
print(mylist)

print(len(mylist))

#Append (you start with the variable first)
mylist.append(666)
print(mylist)

#Insert (you start with the index first and then the inserted element)
mylist.insert( 1 , 4)
print(mylist)

#Remove
Item = mylist.remove(45)
print(mylist)

#Sort
mylist.sort()
print(mylist)
 
#Comparison operators
mylist = [1 , 2 , 3 , 4 , 5]

mylist2 = [6 , 7 , 8 , 9 , 10]

new_list = mylist + mylist2
print(new_list)

# Slicing is a new term that we learn when coming to the knowledge of the python synntax

#...it is when you want to access sub-parts of your list using a ":"

mylist = [1 , 2, 3, 4, 5, 6]

a = mylist[ 2 : 4]
print(a)


# Copying could be used with the slicing function

mylist_org = ["WE ARE WONKRU"]

mylist_copy = mylist_org[:]

#..if you would like to modify the copy from the original :

mylist_copy.append(", BLOOD FOR BLOOD")

print(mylist_copy)
print(mylist_org)

# List Comprehension ( how to create a new list from an existing list using one line.)

#.. the syntax that are used are expressions(i/x) and for in loops 

# E.g square the original least and  create a new list for the elements produced

mylist_a = [2 , 4 , 6 , 8]

mylist_b = [i*i for i in mylist_a]

print(mylist_b)
```
